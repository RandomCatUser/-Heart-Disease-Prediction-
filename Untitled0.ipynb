### Model Testing
Now we will run a test script to verify the model's prediction capabilities.

```python
if os.path.exists('predict.py'):
    print("Running a test prediction...")
    # Running the prediction script which typically handles sample input
    !python predict.py
else:
    print("predict.py not found. Let's create a manual test.")
    import pickle
    import pandas as pd

    # Manual test if script isn't straightforward
    with open('models/heart_model.pkl', 'rb') as f:
        model = pickle.load(f)

    # Creating a dummy sample based on standard heart disease features
    sample_data = pd.DataFrame([[52, 1, 0, 125, 212, 0, 1, 168, 0, 1.0, 2, 2, 3]],
                               columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal'])

    prediction = model.predict(sample_data)
    print(f"Manual Test Prediction for sample: {prediction[0]}")
```
